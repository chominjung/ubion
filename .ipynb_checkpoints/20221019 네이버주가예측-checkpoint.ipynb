```python
import FinanceDataReader as fdr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pmdarima.arima import ndiffs, auto_arima
import pmdarima as pm
```

```python
stocklist_df=fdr.StockListing('KRX')
```

```python
ticker=stocklist_df[stocklist_df['Name']=='NAVER']
```

```python
ns=fdr.DataReader('035420','2020')
ns
```

```python
y_train=ns['Close'][:int(0.7*len(ns))]
y_test=ns['Close'][int(0.7*len(ns)):]
y_train.plot()
y_test.plot()
```

```python
# 차분 차수 구하기
kpss_dfffs=ndiffs(y_train,alpha=0.05, test='kpss',max_d=6)
adf_dfffs=ndiffs(y_train,alpha=0.05, test='adf',max_d=6)
```

```python
max(kpss_dfffs, adf_dfffs)
```

### auto arima
http://alkaline-ml.com/pmdarima/modules/generated/pmdarima.arima.auto_arima.html?highlight=auto_arima

```python
# auto_arima
model=auto_arima(y=y_train
                 ,d=1
                 ,start_p=0
                 ,max_P=3
                 ,start_q=0
                 ,max_q=3
                 ,m=1
                 ,seasonal=False
                 ,stepwise=True
                 ,trace=True
                )
```

### arima(p,d,q)
p는 AR모형 파라미터/
d는 시차 파라미터/
q는 MA모형 파라미터


p=자기회귀 부분의 차수/
d=1차 차분이 포함된 정도/
q=이동 평균 부분의 차수

```python
model.summary()
```

```python
model.plot_diagnostics(figsize=(16,8))
plt.show()
```

### 예측

```python
y_pred=model.predict(n_periods=len(y_test))
y_pred
```

```python
import matplotlib.pyplot as plt
y_pred=model.predict(n_periods=len(y_test))
y_pred=pd.DataFrame(list(y_pred), index=y_test.index, columns=['Prediction'])
y_train=ns['Close'][:int(0.7*len(ns))]
y_test=ns['Close'][int(0.7*len(ns)):]
plt.plot(y_train, label='Train')
plt.plot(y_test, label='Test')
plt.plot(y_pred, label='Prediction')
plt.show()
```

```python
def forecast_one_step():
    fc, conf_int=model.predict(n_periods=1, return_conf_int=True)
    return(fc.tolist()[0], np.asarray(conf_int).tolist()[0])
#     print(result)
forecast_one_step() 
```

```python
forescasts=[]
y_pred=[]
pred_upper=[]
pred_lower=[]

for new_ob in y_test:
    fc, conf=forecast_one_step()
    y_pred.append(fc)
    pred_upper.append(conf[1])
    pred_lower.append(conf[0])
    
    model.update(new_ob)
```

```python
pd.DataFrame({'test':y_test, 'pred':y_pred})
```
